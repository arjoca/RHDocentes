```python
import plotly.express as px
data_canada = px.data.gapminder().query("country == 'Canada'")
fig = px.bar(data_canada, x='year', y='pop')
fig.show()
```

```python
fig.write_image("image.png")
from IPython.display import Image
Image("image.png")
```
